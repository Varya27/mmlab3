```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
x0=32888
y0=17777
t0=0
dt=0.05
tmax=1
```

```python
a1=0.55
b1=0.77
c1=0.66
h1=0.44
```

```python
t=np.arange(t0,tmax,dt)
v0=np.array([x0,y0])
```

```python
def fp1(t):
    p=1.5*math.sin(3*t+1)
    return p
```

```python
def fq1(t):
    q=1.2*math.cos(t+1)
    return q
```

```python
def syst1(y,t):
    dy1=-a1*y[0]-b1*y[1]+fp1(t)
    dy2=-c1*y[0]-h1*y[1]+fq1(t)ч
    return [dy1, dy2]
```

```python
y1=odeint(syst1, v0, t)
```

```python
plt.plot(t,[i[0] for i in y1], lw=2)
plt.plot(t,[i[1] for i in y1], lw=2)
plt.xlabel('время', fontsize=10)
plt.ylabel('численность армии', fontsize=10)
plt.title('Модель боевых действий №1', fontsize=15)
plt.grid(True)
```

```python
a2=0.27
b2=0.88
c2=0.68
h2=0.37
```

```python
def fp2(t):
    p=math.sin(20*t)
    return p
```

```python
def fq2(t):
    q=math.cos(10*t)+1
    return q
```

```python
def syst2(y,t):
    dy1=-a2*y[0]-b2*y[1]+fp2(t)
    dy2=-c2*y[0]*y[1]-h2*y[1]+fq2(t)
    return [dy1, dy2]
```
